```python
import operator

import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures
import pandas as pd

ideal_data = pd.read_csv('data/ideal.csv', index_col=0, delimiter=',')
test_data = pd.read_csv('data/test.csv', index_col=0, delimiter=',')
train_data = pd.read_csv('data/train.csv', index_col=0, delimiter=',')

# Create DataFrames from the CSV files 
ideal_df = pd.DataFrame(ideal_data)
train_df = pd.DataFrame(train_data)
test_df = pd.DataFrame(test_data)
test_df = test_df.sort_index(ascending=False)

ideal_df = ideal_df 
test_df = test_df 
train_df = train_df 

x = train_df.index 
y = train_df['y4']

# transforming the data to include another axis
x = x[:, np.newaxis]
y = y[:, np.newaxis]

polynomial_features= PolynomialFeatures(degree=10)
x_poly = polynomial_features.fit_transform(x)

model = LinearRegression()
model.fit(x_poly, y)
y_poly_pred = model.predict(x_poly)

rmse = np.sqrt(mean_squared_error(y,y_poly_pred))
r2 = r2_score(y,y_poly_pred)
#print(rmse) #MSE
#print(r2)

plt.scatter(x, y, s=10)
# sort the values of x before line plot
sort_axis = operator.itemgetter(0)
sorted_zip = sorted(zip(x,y_poly_pred), key=sort_axis)
x, y_poly_pred = zip(*sorted_zip)
plt.plot(x, y_poly_pred, color='m')
plt.show()

y_poly_pred = model.predict(x_poly)
best = None
lowest = 2
for x in range(1,51):
    rmse = np.sqrt(mean_squared_error(y,y_poly_pred))
    r2 = r2_score(ideal_df['y'+str(x)],y_poly_pred)
    if abs(r2) < lowest:
        lowest = abs(r2)
        best = x
print(best)
```
